```python
train_col_counter = Counter()
length = len(list(TRAIN_DIR.iterdir()))
for ind, file in enumerate(TRAIN_DIR.iterdir()):
    print(f"\r{ind + 1:4}/{length:4}", end="")
    with open(file, "r") as f:
        f.readline()
        age_line = f.readline()
    train_col_counter.update(age_line.strip("# \n").split(","))
```

```python
eval_col_counter = Counter()
length = len(list(EVAL_DIR.iterdir()))
for ind, file in enumerate(EVAL_DIR.iterdir()):
    print(f"\r{ind + 1:4}/{length:4}", end="")
    with open(file, "r") as f:
        f.readline()
        age_line = f.readline()
    eval_col_counter.update(age_line.strip("# \n").split(","))
```

```python
for ind, train_col in enumerate(train_col_counter.most_common()):
    col, value = train_col
    print(f"{col:14}: (train) => {value:4}\t(eval) => {eval_col_counter[col]:4}")
```
