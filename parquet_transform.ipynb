```python
import pandas as pd
from pathlib import Path
```

```python
DATA = Path("data")

CSV = DATA / "csv"
PARQUET = DATA / "parquet"
```

```python
mapper = {"train": {}, "eval": {}}


files = list(CSV.glob("**/*.csv"))
for ind, file in enumerate(files):
    print(f"\r{ind + 1:4}/{len(files):4}", end="")
    with open(file, "r") as f:
        age = f.readline()
        header = f.readline()

    mapper[file.parts[-2]][f"{file.stem}.parquet"] = age.strip("# \n").split(" ")[-1]

    df = pd.read_csv(file, skiprows=[0, 1], engine="c", names=header.strip("# \n").split(","))
    df.to_parquet(PARQUET / file.parts[2] / file.with_suffix(".parquet").name)

pd.DataFrame(mapper["train"].items(), columns=["file", "age"]).to_csv(
    PARQUET / "train_age.csv", index=False
)
pd.DataFrame(mapper["eval"].items(), columns=["file", "age"]).to_csv(PARQUET / "eval_age.csv", index=False)
```
